```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
n = 100
x = np.random.randn(n) # 100 samples

def update(curr):
    # check if animation is at the last frame, and if so, stop the animation a
    if curr == n: 
        a.event_source.stop()
    plt.cla() # clears current axes
    bins = np.arange(-4, 4, 0.5) 
    plt.hist(x[:curr], bins=bins) # plot from beginning to current iteration
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sampling the Normal Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n = {}'.format(curr), [3,27])
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval=100) # (figure, name of fucntion, amt of time we watn in ms between updates

HTML(a.to_jshtml())
```

```python
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Eevent at pixels {}, {} {}and data {},{}'.format(event.x,event.y,'\n',event.xdata,event.ydata))
    
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```

```python
from random import shuffle
origins = ['China', 'Brazil', 'India', 'USA', 'Canada', 'UK', 'Germany', 'Iraq', 'Chile', 'Mexico']

shuffle(origins)

df = pd.DataFrame({'height': np.random.rand(10),
                   'weight': np.random.rand(10),
                   'origin': origins})
df
```

```python
plt.figure()
plt.scatter(df['height'], df['weight'], picker=5) # mouse doesnt hace to clcik directly on rendered object that could be upt to
# 5 pixels away
```

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Selected item came from {}'.format(origin))
    
plt.gcf().canvas.mpl_connect('pick_event', onpick)
```
