```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 2 by 2 grid of subplots
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,sharex=True)
# list containing all subplot figures
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    # sample size = [10,100,1000,10000]
    sample_size = 10**(n+1)
    # creates a normal distribution based on given parameters
    sample_data = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    # creates histogram with sample data generated in nth position of list of subplot figures
    axs[n].hist(sample_data)
    # sets title for the plot
    axs[n].set_title('n={}'.format(sample_size))

# As the sample size increases, the normal distribution created begins to look more normal
# the bar size increases because only the x-axis is shared among all plots
# more sample size also lead to more outliers naturally

# same code as above except with bigger bin size to account for more data to be shown in more detail
# Default number of bins for histograms is 10
# This plot will look much smoother than the above one
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample_data = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample_data, bins=100) # specifies 100 bins/bars
    axs[n].set_title('n={}'.format(sample_size))
    
# How do I know how many bins to specify?
```

```python
# GridSpec allows you to map axes over multiple cells in a grid

plt.figure() # create figure

y = np.random.normal(loc=0,scale=1.0,size=10000) # y -> normal dist
x = np.random.random(size=10000) # x -> random dist
plt.scatter(x,y)

# 3 by 3 grid
# fig, ((ax1,ax2,ax3),(ax4,ax5,ax6),(ax7,ax8,ax9)) = plt.subplots(3,3,sharex=True, sharey=True)
# OR

import matplotlib.gridspec as gridspec # import gridspec library first

plt.figure() # create a figure
gspec = gridspec.GridSpec(3,3) # (rows, columns)

top_hist = plt.subplot(gspec[0,1:]) # takes up row 0, col 1 and on
side_hist = plt.subplot(gspec[1:,0]) # takes rows 1 and on, in column 0
lower_right = plt.subplot(gspec[1:,1:]) # takes rows 1 and on, in columns 1 and on

lower_right.scatter(x,y) # makes a scatter plot in lower right plot of gridspec
side_hist.hist(y, bins=100, orientation='horizontal') # changes orientation of histogram
top_hist.hist(x, bins=100)

# Can clear the plots as well with .clear()
side_hist.clear()
side_hist.hist(y, bins=100, orientation='horizontal', normed=True) # changes orientation of histogram
top_hist.clear()
top_hist.hist(x, bins=100, normed=True)

# invert an axis
# side_hist.invert_xaxis()
```
