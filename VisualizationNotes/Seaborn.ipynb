```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# SEABORN ALLOWS FOR DATA VISUALIZATIONS TO BE MORE APPEALING
# Styles for graphs and charts

np.random.seed(1234)

# size = 1000 , mean = 0, std = 10
v1 = pd.Series(np.random.normal(0,10,1000), name='v1')
v2= pd.Series(2*v1 + np.random.normal(60,15,1000), name='v2')
```

```python
plt.figure()

# specific bins we want to use
plt.hist(v1, alpha=.7, bins=np.arange(-50,150,5), label='v1');
plt.hist(v2, alpha=.7, bins=np.arange(-50,150,5), label='v2');
plt.legend();
```

```python
plt.figure()
plt.hist([v1,v2], histtype='barstacked', normed=True);
v3 = np.concatenate((v1,v2))
sns.kdeplot(v3)
```

```python
plt.figure()
# hist is teal color, kde line is navy color
sns.distplot(v3, hist_kws={'color':'Teal'}, kde_kws={'color':'Navy'})
```

```python
sns.jointplot(v1,v2,alpha=.4);
```
